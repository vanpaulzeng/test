```python
#===============================================This is a comment

"""=============================================This is a comment written in more than just one line
Hello I need to add more commets
"""



#===============================================Simple for loop
fruits = ["apple", "banana", "cherry", "pear"]
for x in fruits:
  print(x)

#===============================================This is for variables
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0

print(x,y,z)
print(x)
print(y)

#==================================Assign multiple values
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)

#Globle Values: different x will be shown, 1st x is inside func only, 2nd x is globle
x = "awesome"

def myfunc():
  x = "fantastic"
  global y 
  y="this is global inside func"
  print("Python is " + x)

myfunc()

print("Python is " + x + y)


#=============================================Data types
x = "Hello World" #str 
x = 20  #int 
x = 20.5  #float 
x = 1j  #complex 
x = ["apple", "banana", "cherry"] #list, List items are ordered, changeable, and allow duplicate values.
x = ("apple", "banana", "cherry") #tuple, A tuple is a collection which is ordered and unchangeable.
x = range(6)  #range 
x = {"name" : "John", "age" : 36} #dict,  A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
x = {"apple", "banana", "cherry"} #set, A set is a collection which is unordered, unchangeable*, and unindexed.
x = frozenset({"apple", "banana", "cherry"})  #frozenset 
x = True  #bool  
x = b"Hello"  #bytes 
x = bytearray(5)  #bytearray 
x = memoryview(bytes(5))  #memoryview  
x = None #NoneType

#======================================================== String value

a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididuntut labore et dolore magna aliqua."""

b = "One line only"
print(a)
print(b)


#======================================================= Slice string

a = "  Hello,  World!    "
print(a[1:5])
print(a.replace("H", "J"))
print(a.upper())
print(a.lower())
print(a.lower())
print(a.lower())
print(a.strip())
print(a.split(",")) # returns ['Hello', ' World!']

#======================================================= IF elif
a = 31
b = 34
if b > a:
  print("b is greater than a")
  print("another line")
elif a == b:
  print("a and b are equal")
else:
  print("a is great than b")
  

#========================================================= While loop
i = 1
while i < 6:
  print(i)
  if i == 3:
    break # or continue
  i += 1

#========================================================= For loop
i = 1
while i < 5:
   print(i)
   i += 1
   print("aaa")  

#========================================================= Math
import math

x = math.sqrt(64)
print(x)

x = pow(4, 3)
print(x)

import json


#========================================================= some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'  #dict

# parse x:
y = json.loads(x)

# the result is a Python dictionary:
print(y["age"],":",y["city"])

#========================================================= input:
#username = input("Enter username:")
#print("Username is: " + username)


#========================================================= Date:
import datetime

x = datetime.datetime.now()

print (x)
print(x.year)
print(x.strftime("%A"))


#=========================================================  Program to display calendar of the given month and year

# importing calendar module
import calendar

yy = 2023  # year
mm = 11    # month

# To take month and year input from the user
# yy = int(input("Enter year: "))
# mm = int(input("Enter month: "))

# display the calendar
print(calendar.month(yy, mm))


#=========================================================  Python program to shuffle a deck of card

# importing modules
import itertools, random

# make a deck of cards
deck = list(itertools.product(range(1,14),['Spade','Heart','Diamond','Club']))

# shuffle the cards
random.shuffle(deck)

# draw five cards
print("You got:")
for i in range(5):
   print(deck[i][0], "of", deck[i][1])


#========================================================= 
"""

Base Python has only a few math operations
x + y sum of x and y.
x * y multiplication of x and y.
x - y difference of x and y.
x / y division of x by y.
x // y integer floor division of x by y.
x % y integer remainder of x//y
x ** y x to the power of y
abs(x) absolute value of x
"""


#========================================================= File Handling
f = open("Demo.txt", "r")
print(f.readline())
print(f.readline(10))

f = open("Demo.txt", "r")
for x in f:
  print(x)

f.close()

f = open("Demo.txt", "a")
f.write("\n Now the file has more content!")
f.close()

#========================================================= 
import matplotlib

print(matplotlib.__version__)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 25])

plt.plot(xpoints, ypoints)
plt.show()


y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()


x = np.array([80, 125])
y = np.array([240, 200])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.show()


#========================================================= Machine Learning

"""
Hello I need to add more commets
Hello I need to add more commets

To analyze data, it is important to know what type of data we are dealing with.
That is what Machine Learning is for! Analyzing data and predicting the outcome!

We can split the data types into three main categories:

Numerical
Categorical
Ordinal

Numerical data are numbers, and can be split into two numerical categories:

Discrete Data
- numbers that are limited to integers. Example: The number of cars passing by.
Continuous Data
- numbers that are of infinite value. Example: The price of an item, or the size of an item
Categorical data are values that cannot be measured up against each other. Example: a color value, or any yes/no values.

Ordinal data are like categorical data, but can be measured up against each other. Example: school grades where A is better than B and so on.

By knowing the data type of your data source, you will be able to know what technique to use when analyzing them.



Mean - The average value
Median - The mid point value
Mode - The most common value

"""


import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.mean(speed)
y = numpy.median(speed)

from scipy import stats
z = stats.mode(speed)

print(x)
print(y)
print(z)




#========================================================= 机器学习
#========================================================= 
#========================================================= 
#========================================================= 
#========================================================= 
#========================================================= 



```
